# FUNCIONA. SACO COLUMNA CELDAS CON TODO

```python
driver = webdriver.Chrome()
print("abro navegador")
sleep(2)
driver.maximize_window()
print("maximizo ventana")
sleep(2)
driver.get("https://www.vinissimus.com/es/")
print("entro a la vinissimus ")
#Aceptamos las cookies
sleep(2)
driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/button[2]").click()
print("acepto cookies")
sleep(2)
#Accedemos a Vinos 
driver.find_element(By.XPATH,"/html/body/nav/ul/li[1]/a").click()
print("accedo a vinos")
sleep(2)
#Accedemos a españoles
driver.find_element(By.XPATH,"/html/body/div/main/div/section/div[1]/button[2]").click()
print("accedo a españoles")
sleep(2)


dict_results = {"name_w": [], "celdas": []}

# Tu bucle para iterar entre páginas
for page in range(1, 30):  # ITERACIÓN DE PÁGINAS
    
    try: 
        # Hacer clic en el botón de la página actual
        driver.find_element(By.XPATH, f'//*[@id="__next"]/main/div[2]/section/nav/button[{page}]').click()
    
    except NoSuchElementException:
        print(f"Botón de la página {page} no encontrado. ¡Continuando con la siguiente página!")
        continue

    sleep(5)



    for wine in range(2, 25): 
        try:

            wine_element = driver.find_element(By.CSS_SELECTOR, f'#__next > main > div.content.theme.clear > section > div.product-list > div > div:nth-child({wine}) > div.info > div.details > a')
            actions = ActionChains(driver)
            actions.move_to_element(wine_element).click().perform()
            wine_element.click()
            print("Accediendo al vino")
            sleep(2)
        except NoSuchElementException:
            print(f"No se encontró el elemento para el vino {wine}. Saltando...")
            continue


        # --------- var NAME ------------- 
        try:
            name_w= driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[2]/section/div[1]/div[2]/div/div[1]/div/h1').text
            dict_results['name_w'].append(name_w)
            print(name_w)
        except NoSuchElementException:
            dict_results['name_w'].append('unknown')

        # SCROLL
        driver.execute_script("window.scrollTo(0, 2000)")
        print("Haciendo scroll")
        sleep(3)
            
        try: 
        # Encontrar la tabla por su clase
            tabla = driver.find_element(By.CSS_SELECTOR, '.styles_table__jWPKo.info-grid.score-awards')

        
            #dict_results['tabla'].append(tabla.get_attribute('outerHTML'))  # Almacenar el HTML de la tabla

            # Encontrar las filas de la tabla
            filas = tabla.find_elements(By.TAG_NAME, 'tr')
            #dict_results['filas'].append([fila.text for fila in filas])  # Almacenar texto de las filas

            # Iterar sobre las filas e imprimir el texto de cada celda
            celdas_texto = []
            for fila in filas:
                celdas = fila.find_elements(By.TAG_NAME, 'td')  # Cambia "td" por "th" si son celdas de encabezado
                for celda in celdas:
                    celdas_texto.append(celda.text)  # Almacenar texto de las celdas
            dict_results['celdas'].append(celdas_texto)  # Agregar a la lista de celdas
        
        except NoSuchElementException:
            # Manejar la situación en la que no se encuentra la tabla
            print("La tabla no se encontró. Realizar acciones necesarias.")
            # Puedes agregar aquí cualquier otro código para manejar esta situación
        driver.back()
        sleep(6)

driver.close()
print(dict_results)
```

# AQUI HACEMOS EXTRACCION DE PRIMER VALOR DE CADA COLUMNA DE PUNTOS 

```python
import pandas as pd

# Suponiendo que df_wines es tu DataFrame y "celdas" es la columna específica que contiene las celdas
# Puedes reemplazar 'celdas' con el nombre real de tu columna si es diferente

def extract_values(row):
    pk = ''
    pn = ''
    sk = ''
    for val in row:
        if 'PK' in val:
            pk = val.split()[0]
        elif 'PN' in val:
            pn = val.split()[0]
        elif 'SK' in val:
            sk = val.split()[0]
    return pd.Series({'PK': pk, 'PN': pn, 'SK': sk})

# Aplicar la función a la columna 'celdas' y concatenar los resultados al DataFrame original
df_extracted = df_wines["celdas"].apply(extract_values)
df_wines[['PK', 'PN', 'SK']] = df_extracted[['PK', 'PN', 'SK']]

# Mostrar el DataFrame resultante
print(df_wines)




```
