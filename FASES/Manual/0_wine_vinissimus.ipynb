```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from time import sleep
import pandas as pd 
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Chrome()
print("abro navegador")
sleep(2)
driver.maximize_window()
print("maximizo ventana")
sleep(2)
driver.get("https://www.vinissimus.com/es/")
print("entro a la vinissimus ")
#Aceptamos las cookies
sleep(2)
driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/button[2]").click()
print("acepto cookies")
sleep(2)
#Accedemos a Vinos 
driver.find_element(By.XPATH,"/html/body/nav/ul/li[1]/a").click()
print("accedo a vinos")
sleep(2)
#Accedemos a españoles
driver.find_element(By.XPATH,"/html/body/div/main/div/section/div[1]/button[2]").click()
print("accedo a españoles")
sleep(2)
#Entro al primer vino 

#Recojo la info de la ficha


dict_results = {
    "name_w": [], "year_w": [], "price": [], "user_score": [], "type_w": [], "region": [], 
    "producer": [], "alcohol_content": [], "wine_grape": [], "pk_score": [], "pn_score": [],
    "sk_score":[]
}



# Tu bucle para iterar entre páginas
for page in range(1, 30):  # ITERACIÓN DE PÁGINAS
    
    try: 
        # Hacer clic en el botón de la página actual
        driver.find_element(By.XPATH, f'//*[@id="__next"]/main/div[2]/section/nav/button[{page}]').click()
    
    except NoSuchElementException:
        print(f"Botón de la página {page} no encontrado. ¡Continuando con la siguiente página!")
        continue

    sleep(5)



    for wine in range(2, 25): 
        try:

            wine_element = driver.find_element(By.CSS_SELECTOR, f'#__next > main > div.content.theme.clear > section > div.product-list > div > div:nth-child({wine}) > div.info > div.details > a').get_attribute("href")
            actions = ActionChains(driver)
            actions.move_to_element(wine_element).click().perform()
            wine_element.click()
            print("Accediendo al vino")
            sleep(2)
        except NoSuchElementException:
            print(f"No se encontró el elemento para el vino {wine}. Saltando...")
            continue


        # --------- var NAME ------------- 
        try:
            name_w= driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[2]/section/div[1]/div[2]/div/div[1]/div/h1').text
            dict_results['name_w'].append(name_w)
            print(name_w)
        except NoSuchElementException:
            dict_results['name_w'].append('unknown')



        # --------- var YEAR -------------
        try: 
            year_w = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[2]/section/div[1]/div[2]/div/div[1]/div/span').text
            dict_results['year_w'].append(year_w)
            print(year_w)
        except NoSuchElementException:
            dict_results['year_w'].append('unknown')
        

        
        # --------- var PRICE -------------
            
        try: 
            price = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[2]/section/div[2]/div[2]/p').text
            dict_results['price'].append(price)
            print(price)
        except NoSuchElementException:
            dict_results['price'].append('unknown')


        
        # ---------- var type of Wine ---------------- FIXED 
        sleep(2)
        type_w = 'unknown'
        
        try: 
            type_w = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[4]/section/div[1]/div/table/tbody/tr[1]/td/a').text
            
            
        except NoSuchElementException:
            try:
                type_w = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[5]/section/div[1]/div/table/tbody/tr[1]/td/a').text

            except NoSuchElementException:
                pass

        dict_results['type_w'].append(type_w)
        print(type_w)


        # ---------- var REGION ------------------- FIXED 
        sleep(2)
        region = 'unknown'  # Valor por defecto si no se encuentra la información en ninguno de los XPaths

        try:
            region_element = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[5]/section/div[1]/div/table/tbody/tr[2]/td/div/a[1]')
            region = region_element.text
        except NoSuchElementException:
            try:
                region_element = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[4]/section/div[1]/div/table/tbody/tr[2]/td/div/a[1]')
                region = region_element.text
            except NoSuchElementException:
                pass  # Si no se encuentra en ninguno de los XPaths, se mantiene el valor 'unknown'

        dict_results['region'].append(region)
        print(region)

        # ----- SCROLL -----------------------

        driver.execute_script("window.scrollTo(0, 1400)")
        print("hago scroll")
        sleep(2)

        # ---------- var PRODUCER ---------------- FIXED
        sleep(2)
        producer = 'unknown'
        
        try: 
            producer = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[5]/section/div[1]/div/table/tbody/tr[4]/td/div').text
            
        except NoSuchElementException:
            try:
                producer = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[4]/section/div[1]/div/table/tbody/tr[4]/td/div/a').text

            except NoSuchElementException:
                pass

        dict_results['producer'].append(producer)
        print(producer)

        # --------- var ALCOHOL CONTENT ------------- FIXED 
        sleep(4)
        alcohol_content = 'unknown'
        
        try: 
            alcohol_content = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[4]/section/div[1]/div/table/tbody/tr[6]/td').text

            
        except NoSuchElementException:
            try:
                alcohol_content = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[6]/section/div[1]/div/table/tbody/tr[6]/td ').text

            except NoSuchElementException:
                pass

        dict_results['alcohol_content'].append(alcohol_content)
        print(alcohol_content)

        
        # --------- METER POP UP DEL DESCUENTO AQUI  -------------
        # ---------- var wine grape ----------------  FIXED! 
        sleep(5)
        wine_grape = 'unknown'
        
        try: 
            wine_grape = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[5]/section/div[1]/div/table/tbody/tr[3]/td/div').text 
        except NoSuchElementException:
            try:
                wine_grape = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[4]/section/div[1]/div/table/tbody/tr[3]/td/div').text

            except NoSuchElementException:
                pass

        dict_results['wine_grape'].append(wine_grape)
        print(wine_grape)
        
        # ----- SCROLL -----------------------

        driver.execute_script("window.scrollTo(0, 2400)")
        print("hago scroll")
        sleep(2)
        
                # --------- var USER SCORE ------------- 

        try: 
            user_score = driver.find_element(By.CLASS_NAME, 'styles_grade__EU5A1').text
            dict_results['user_score'].append(user_score)
            print(user_score)
        
        except NoSuchElementException:
            dict_results['user_score'].append('unknown')
        sleep(2)
        
        # ------ var PK (parker score) ---------------
        
        try:

            pk_score = driver.find_element(By.CSS_SELECTOR, '[title="Parker"]').text
            dict_results['pk_score'].append(pk_score)        
            print(pk_score)
        except NoSuchElementException:
            dict_results['pk_score'].append('unknown')
        sleep(4)


        # ------ var PN (Peñin score) ---------------
        
        try:     
            #pn_score = driver.find_element(By.CSS_SELECTOR,'#__next > main > div:nth-child(6) > section > div:nth-child(1) > div > table > tbody > tr:nth-child(2) > td:nth-child(3)').text
            pn_score = driver.find_element(By.CSS_SELECTOR, '[title="Peñín"]').text
            dict_results['pn_score'].append(pn_score)  
            print(pn_score)

        except NoSuchElementException:
            dict_results['pn_score'].append('unknown')
        sleep(4)


        # ------ var SK (Suckling score) ---------------
        try:
            #sk_score = driver.find_element(By.CSS_SELECTOR, '#__next > main > div:nth-child(6) > section > div:nth-child(1) > div > table > tbody > tr:nth-child(2) > td:nth-child(4)').text
            sk_score = driver.find_element(By.CSS_SELECTOR, '[title="Suckling"]').text
            dict_results['sk_score'].append(sk_score) 
            print(sk_score)
        except NoSuchElementException:
            dict_results['sk_score'].append('unknown')
        sleep(3)

        # #Go back to the menu 
        print(dict_results["name_w"])
        driver.back()
        sleep(6)

    #Close Chrome 
driver.close()
print(dict_results)

```
