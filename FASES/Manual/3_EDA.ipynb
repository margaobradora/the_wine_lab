```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import requests as re
```

```python
df = pd.read_csv("../Manual/Files/wines_table.csv", index_col=0)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#cambiar las comas por puntos: 

columnas_a_actualizar = ['price', 'user_score', 'alcohol_content']

for columna in columnas_a_actualizar:
    df[columna] = df[columna].astype(str).str.replace(',', '.')
```

```python
#quitar de la var año, la info extra más allá del propio numero 
df['year_w'] = df['year_w'].str.replace(r'\(.*\)', '', regex=True)
```

```python
df
```

```python
#eliminar las letras y dejar solo el numero 
df['pk_score'] = df['pk_score'].str.split().str[0]
df['pn_score'] = df['pn_score'].str.split().str[0]
df['sk_score'] = df['sk_score'].str.split().str[0]
```

```python
# quitar el € de price y el % de alcohol content.v

df['price'] = df['price'].str.replace('€', '')
df['alcohol_content'] = df['alcohol_content'].str.replace('%', '')
```

```python
df['price'].isna().sum()
```

```python
# pasar price, user_score, alcohol_content a float 

def convert_to_float(val):
    try:
        return float(val)
    except (ValueError, TypeError):
        return None  # Or any other value to indicate non-convertible strings

df['price'] = df['price'].apply(convert_to_float)
df['user_score'] = df['user_score'].apply(convert_to_float)
df['alcohol_content'] = df['alcohol_content'].apply(convert_to_float)

# pasar las 3 de scoring a int
```

```python
df.dtypes
```

```python
# pasar las 3 de scoring a int

def convert_to_integer(val):
    try:
        return int(val)
    except (ValueError, TypeError):
        return None  # Or any other value to indicate non-convertible strings

# Applying the function to convert 'price' column
df['pk_score'] = df['pk_score'].apply(convert_to_integer)
df['pn_score'] = df['pn_score'].apply(convert_to_integer)
df['sk_score'] = df['sk_score'].apply(convert_to_integer)

```

```python
df
```

```python
# Suponiendo que 'final_df' es tu DataFrame y quieres verificar duplicados en la columna 'name_w'

duplicates = df[df.duplicated(subset=['name_w'], keep=False)]
duplicates.head(2)
```

```python
# Suponiendo que 'duplicates' es el DataFrame con las filas duplicadas basadas en 'name_w'

# Ordenar por 'name_w' en orden alfabético
duplicates_sorted = duplicates.sort_values(by='name_w', ascending=True)
duplicates_sorted
```

```python
# Suponiendo que 'final_df' es tu DataFrame y quieres eliminar duplicados basados en 'name_w'

# Eliminar duplicados basados en 'name_w' y conservar el primero
df = df.drop_duplicates(subset=['name_w'], keep='first')
```
